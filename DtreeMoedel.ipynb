```python
import pandas as pd
```

# Load the the data set into the dataframe for both white and Red Wine 

```python
tr_red = pd.read_csv('resources/winequality-red.csv', sep=';')
tr_white = pd.read_csv('resources/winequality-white.csv', sep=';')
frames=[tr_red,tr_white]
tr_feature= pd.concat(frames)
tr_feature.head()
```

# Define a  Function To Categorize The Quality 
1. If Quality is <=4 then Low Quality
2. If Quality is <4 and <= 7 then Medium
3. If Quality is >7 Then High

```python
def bucket(value):
    if value <= 4:
        return 'low'
    elif value > 4 and value <= 7:
        return 'medium'
    else:
        return 'high'

tr_feature['label'] = tr_feature['quality'].apply(lambda x: bucket(x))
```

# Create Label and Parameter 

Divide the data into Train and Test 

```python
data = tr_feature.values
X = data[:, 0:11]
y = data[:, 12]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# Create the Model based on D Tree And Save  The Model

```python
# from sklearn.preprocessing import LabelEncoder
# from sklearn import tree
# import joblib
# # Step 1: Label-encode data set
# label_encoder = LabelEncoder()
# label_encoder.fit(y)
# encoded_y = label_encoder.transform(y)


# clf = tree.DecisionTreeClassifier()
# clf = clf.fit(X, y)
# joblib.dump(clf, './Dtree_model.pkl')

from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import joblib
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y = label_encoder.transform(y)


clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
joblib.dump(clf, './Dtree_model.pkl')
```

# Check the Score of The Model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```
